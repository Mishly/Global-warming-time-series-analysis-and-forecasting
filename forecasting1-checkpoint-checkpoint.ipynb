Key Components in Seasonal Decomposition:
seasonal_decompose splits your time series into 3 main parts:
Trend 📈
The long-term progression (e.g., overall increase in temperature due to global warming).
Seasonality 🔄
Repeating patterns at regular intervals (e.g., temperature rising in summer and falling in winter).
period=7 means you’re assuming weekly seasonality (e.g., daily data with patterns repeating every 7 days).
Residual (Noise) ❓
What's left after removing trend and seasonality (the unpredictable part).

The ADF test is used to check if your time series is ready for modeling. If it's not stationary, your model won't perform well — so it's a key preprocessing step for accurate forecasting, including for global warming trends.

```python
from statsmodels.tsa.stattools import adfuller
def ad_test(dataset):
    dftest = adfuller(dataset,autolag = 'AIC')
    print("1. ADF :", dftest[0])
    print("2. P.value :", dftest[1])
    print("3. Num of lags :", dftest[2]) 
    print("4. Num of observations used for ADF Regression and critical value calculation :", dftest[3])
    print("5. Critical values :")
    for key,val in dftest[4].items():
      print("\t",key,":",val)
```

When you run ad_test(df['DAYTON_MW']) or similar:
ADF Statistic: A test statistic. More negative = more likely to be stationary.
P-value: The most important one.
If p-value < 0.05, the series is likely stationary (✅ good for modeling).
If p-value ≥ 0.05, the series is likely non-stationary (⚠️ you may need to difference the data).

```python
ad_test(df['DAYTON_MW'])
```

Results show the p value is greater than 0.05 so we hvw to do first differencing to make the data stationary

MODEL DEPLOYMENT
We'll deploy the model using streamlit application.

```python
import streamlit as st
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

st.title('Temperature Forecasting App')

# Upload CSV File
uploaded_file = st.file_uploader("Upload your temperature CSV file", type=["csv"])

if uploaded_file:
    # Load data
    df = pd.read_csv(uploaded_file)
    
    # Show preview
    st.subheader("Preview of Uploaded Data")
    st.write(df.head())

    # Ensure datetime index
    if 'Datetime' in df.columns:
        df['Datetime'] = pd.to_datetime(df['Datetime'])
        df.set_index('Datetime', inplace=True)

    # Check for required column
    if 'DAYTON_MW' not in df.columns:
        st.error("Uploaded CSV must contain a 'DAYTON_MW' column.")
    else:
        # Forecasting functions
        def forecast_arima(data, steps):
            model = ARIMA(data['DAYTON_MW'], order=(5, 1, 0))
            model_fit = model.fit()
            forecast = model_fit.forecast(steps)
            return forecast

        def forecast_sarima(data, steps):
            model = SARIMAX(data['DAYTON_MW'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
            model_fit = model.fit()
            forecast = model_fit.forecast(steps)
            return forecast

        def moving_average(data, window_size):
            return data['DAYTON_MW'].rolling(window=window_size).mean().iloc[-1]

        # Sidebar parameters
        st.sidebar.header('User Input Parameters')
        model_type = st.sidebar.selectbox('Select Model:', ('ARIMA', 'SARIMA', 'Moving Average'))
        forecast_steps = st.sidebar.number_input('Forecast Steps:', min_value=1, max_value=30)

        if model_type == 'Moving Average':
            window_size = st.sidebar.number_input('Window Size:', min_value=1)
        else:
            window_size = None

        if st.button('Run Forecast'):
            if model_type == 'ARIMA':
                forecast_result = forecast_arima(df[-365:], forecast_steps)
                st.write(f"ARIMA Forecast for next {forecast_steps} steps:")

            elif model_type == 'SARIMA':
                forecast_result = forecast_sarima(df[-365:], forecast_steps)
                st.write(f"SARIMA Forecast for next {forecast_steps} steps:")

            elif model_type == 'Moving Average':
                last_temp_ma = moving_average(df[-365:], window_size)
                future_temps_ma = [last_temp_ma] * forecast_steps
                forecast_result = pd.Series(future_temps_ma)

            # Display results
            st.subheader("Forecasted Values")
            st.line_chart(forecast_result)

else:
    st.info("Please upload a CSV file containing temperature data.")
```
